## Createstubs for Micropython-Windows-port
- Needs to be run on windows 
- firmware for windows can be built on WSL using mpbuild or the `build_sa_ports.py` script
- firmware must be accessible from the windows machine ( copied to the windows drive )


```python
# createstubs for the Windows port
# - port,board,variant = windows,None,standard 
# todo : add board / port.py 
# - stubber merge expects - port or GENERIC - ( todo : Add standard variant to stubber merge )

firmware = "C:\\Users\\josverl\\Downloads\\firmware"
dest = "D:\\mypython\micropython-stubber\\repos\\micropython-stubs"
createstubs_py = "D:\\mypython\\micropython-stubber\\src\\stubber\\board\\createstubs.py"

for version in ["v1.24.1", "v1.25.0-preview"]: 
    # run createstubs.py 
    # workaround : set BOARDNAME to "GENERIC" using boardname.py
    !{firmware}\windows\windows-{version}.exe {createstubs_py} --path {dest}
    # merge and build
    !stubber merge --version {version} --port windows
    !stubber build --version {version} --port window

# Publish package to pypi 
publish = False
if publish: 
    !stubber publish --version stable --port window --pypi

# update release json 
# push to github
```
